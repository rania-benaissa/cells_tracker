```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train_results_all = pd.read_pickle("train_results_all.pkl")
train_results1 = pd.read_pickle("train_results1.pkl")
train_results3 = pd.read_pickle("train_results3.pkl")
train_results = pd.read_pickle("train_results.pkl")
```

```python
def plot_map(data):

    ax = plt.gca() 

    x_values = np.arange(0.5, 1.0, 0.05)



    names = data.columns

    data['Model'] = x_values

    for i in range(1,len(names)):

        data.plot(kind = 'line',
        x =names[0],
        y = names[i],ax = ax)

        print(names[i],data[names[i]].mean())

    ax.set_xlabel("IoU")
    ax.set_ylabel("AP")
  

    
```

```python
# this is for strategie 1
plot_map(train_results_all)
```

```python
# this is for strategie 2

data ={}

aps_range = ["map@" + str(ap) for ap in  np.arange(0.5, 1.0, 0.05) ]

data = {'Model': aps_range}

data["cell20230119T0135"] = train_results1["cell20230119T0135"] # model 13
data["cell20230119T1105"] = train_results1["cell20230119T1105"] # model 14
# results train missing model 15 and model 16

data["cell20230118T1114"] = train_results["cell20230118T1114"] # model 15
data["cell20230118T2126"] = train_results["cell20230118T2126"] # model 16

data["cell20230124T1247"] = train_results3["cell20230124T1247"] # model 17

startegy2 = pd.DataFrame(data)

plot_map(startegy2)
```
